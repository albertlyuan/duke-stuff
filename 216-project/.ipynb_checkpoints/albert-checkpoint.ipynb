```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
TeamAdvanced = pd.read_csv("data/TeamAdvancedStats2000_2021.csv")
print("TeamAdvanced")
print(TeamAdvanced.head())
TeamPerGame = pd.read_csv("data/TeamPerGameStats2000_2021.csv")
print("TeamPerGame")
print(TeamPerGame.head())
TeamShooting = pd.read_csv("data/TeamShootingStats2000_2021.csv")
print("TeamShooting")
print(TeamShooting.head())

combinedTeamStats = pd.merge(TeamAdvanced, TeamShooting,  how='left', left_on=['Year','Team'], right_on = ['Year','Team'])
combinedTeamStats = pd.merge(combinedTeamStats, TeamPerGame,  how='left', left_on=['Year','Team'], right_on = ['Year','Team'])
# print("--combined--")
print(combinedTeamStats.columns)
# print(combinedTeamStats.head())
```

```python
# print(combinedTeamStats.columns)
pointsVS3pa = combinedTeamStats[['Year','Team','3PA','PTS','3PAr','SRS']]

plot = sns.relplot(data=pointsVS3pa,x='3PAr',y='PTS').set(title='Percentage of Shots From 3-point Line Compared to Total Points')

x = pointsVS3pa['3PAr'].values.reshape(-1, 1)
y = pointsVS3pa['PTS'].values.reshape(-1, 1)
linReg = LinearRegression()
linReg.fit(x, y)
print("r2: ", linReg.score(x,y))
```

```python
#3pt analysis
SRSvs3PT = combinedTeamStats[['Year','Team','SRS','W','L','3P','3P%','3PAr']]
decade1 = SRSvs3PT[SRSvs3PT['Year'] <= 2010]
decade1['Decade'] = '2000s'
decade2 =SRSvs3PT[SRSvs3PT['Year'] > 2010]
decade2['Decade'] = '2010s'

combinedDecade = pd.concat([decade1,decade2])
# combinedDecade['tag'] = np.where(combinedDecade["Team"]=='Houston Rockets', "Houston","none")
combinedDecade['tag'] = np.where(combinedDecade["Year"]>=2015, "AfterSteph", 'BeforeSteph')

# sns.relplot(data=combinedDecade,x='3PAr',y='SRS',hue='afterSteph',alpha=0.75)
sns.relplot(data=combinedDecade,x='Year',y='3PAr',hue='tag',alpha=0.75).set(title='Percentage of Shots From 3-point Line Before and After Steph Curry')
```

hypothesis test: did the mean proportion of 3 pointers attempted increase after the Warriors' 2015 championship season?

null hypothesis: $\mu_{before} = \mu_{after}$


alternative hypothesis: $\mu_{before} < \mu_{after}$


Because the p value is 2.017e-80, the probability of observing the sample statistic is 2.017e-80, which is less than $\alpha = 0.05$. Thus, we can reject the null hypothesis that the mean proportion of 3 pointers attempted after the Warriors 2015 championship year did not increase. We can conclude that there is significant evidence that the mean proportion of 3 pointers attempted increased after the Warriors 2015 championship year .

```python
preWarriors = pointsVS3pa[(pointsVS3pa['Year']<2015)]
warriorsEra = pointsVS3pa[(pointsVS3pa['Year']>=2015)]
preWarriors['tag'] = 'preWarriors'
warriorsEra['tag'] = 'warriors'

BeforeAfterWarriors= pd.concat([preWarriors, warriorsEra])
sns.catplot(data=BeforeAfterWarriors,y='3PAr',x='tag',kind="box").set(title='Percentage of Shots From 3-point Line Before and After Warriors Era')


p = stats.ttest_ind(preWarriors['3PAr'],warriorsEra['3PAr']).pvalue / 2
print(p)
```

```python
SRSvs3PAr = combinedTeamStats[['Year','Team','3PAr','SRS']].sort_values('3PAr',ascending=False)
SRSvs3PAr['after2015'] = np.where(SRSvs3PAr['Year']>=2015,True,False)
# print(SRSvs3PAr.head(10))
SRSvs3PAr_plot = sns.relplot(data=SRSvs3PAr,x='3PAr',y='SRS',hue='after2015').set(title='Percentage of Shots From 3-point Line Compared to SRS')

overall_x = SRSvs3PAr['3PAr'].values.reshape(-1, 1)
overall_y = SRSvs3PAr['SRS'].values.reshape(-1, 1)
linReg = LinearRegression()
linReg.fit(overall_x, overall_y)
print("2000-2021 r2: ", linReg.score(overall_x,overall_y))

before_x = SRSvs3PAr[SRSvs3PAr['after2015']==False]['3PAr'].values.reshape(-1, 1)
before_y = SRSvs3PAr[SRSvs3PAr['after2015']==False]['SRS'].values.reshape(-1, 1)
linReg = LinearRegression()
linReg.fit(before_x, before_y)
print("before2015 r2: ", linReg.score(before_x,before_y))

after_x = SRSvs3PAr[SRSvs3PAr['after2015']==True]['3PAr'].values.reshape(-1, 1)
after_y = SRSvs3PAr[SRSvs3PAr['after2015']==True]['SRS'].values.reshape(-1, 1)
linReg = LinearRegression()
linReg.fit(after_x, after_y)
print("after2015 r2: ", linReg.score(after_x,after_y))
```
